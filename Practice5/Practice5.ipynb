5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class Figure3D:
  def get_volume(self):
    return 0

  def __add__(self, other):
    return self.get_volume() + other.get_volume()
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math

class Cylinder(Figure3D):
  def __init__(self, r, h):
    self.r = r
    self.h = h

  def get_volume(self):
    return math.pi * (self.r**2) * self.h

class Cube(Figure3D):
  def __init__(self, a):
    self.a = a

  def get_volume(self):
    return self.a**3

cylinder = Cylinder(5,10)
print("Об'єм циліндра",cylinder.get_volume())

cube = Cube(3)
print("Об'єм куба",cube.get_volume())

add = cylinder + cube
print("Сумарний об'єм:",add)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
  def __init__(self, balance, id):
    self._balance = balance
    self._id = id

  @property
  def balance(self):
    return self._balance

  @balance.setter
  def balance(self, balance):
    self._balance = balance

  @property
  def id(self):
    return self._id

  @id.setter
  def id(self, id):
    self._id = id

  def increase_money(self, amount):
    if amount > 0:
      self.balance += amount
    else: print("Error")

  def decrease_money(self, amount):
    if amount > 0 and amount <= self.balance:
      self.balance -= amount
    else: print("Error")

wallet1 = EWallet(200,1)
print("Грошей на гаманці",wallet1.balance)
wallet1.increase_money(100)
print("Грошей після поповнення",wallet1.balance)
wallet1.decrease_money(50)
print("Грошей після зняття",wallet1.balance)
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
from typing import List

class Item:
  def __init__(self, name:str):
    self.name = name

  def __str__(self) -> str:
    return self.name

class ShoppingCart:
  def __init__(self, items: List[Item]):
    self.items = items

  def add_item(self, item: Item) -> None:
    self.items.append(item)

  def delete_item(self, item_name: str) -> None:
    for item in self.items:
      if item.name == item_name:
        self.items.remove(item)
        break

  def __repr__(self) -> str:
    if not self.items:
      return ("Shopping Cart порожній")
    res = "Товари в Shopping Cart:\n"
    for item in self.items:
      res += f"{str(item)}\n"
    return res

item1 = Item('product1')
item2 = Item('product2')
item3 = Item('product3')
item4 = Item('product4')
items = list((item1, item2, item3, item4))
cart = ShoppingCart(items)
print(cart)

cart.add_item(Item('product5'))
print("Після додавання product5")
print(cart)

cart.delete_item(Item('product5'))
print("Після видалення product5")
print(cart)
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class DayOfWeek:
  def __init__(self):
    self.day = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    self.index = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.index < len(self.day):
      day_week = self.day[self.index]
      self.index += 1
      return day_week
    else: raise StopIteration

week = DayOfWeek()
print(next(week))
print(next(week))
print(next(week))
print(next(week))
print(next(week))
print(next(week))
print(next(week))
#print(next(week))
```
